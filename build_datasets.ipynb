```python
import numpy as np
import pandas as pd
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
def copy_files(src, dest):
    if not os.path.exists(dest):
        os.makedirs(dest)
    src_files = os.listdir(src)
    for file_name in src_files:
        full_file_name = os.path.join(src, file_name)
        if os.path.isfile(full_file_name):
            shutil.copy(full_file_name, dest)
def find_sepsis_file(data_path_dir,id_nosepsis,id_sepsis):
    # id_nosepsis = []
    # id_sepsis = []
    for psv in os.listdir(data_path_dir):
        pid = pd.read_csv(os.path.join(data_path_dir, psv), sep='|')
        if 'SepsisLabel' not in pid.columns:
            pass
        else:
            if 1 in np.array(pid.SepsisLabel):
                id_sepsis.append(psv)
            else:
                id_nosepsis.append(psv)
    return (id_nosepsis, id_sepsis)
```

```python
hours_added_path = '/content/drive/MyDrive/cpbandit/CP_Bandit/data/hours2sepsis/'
if not os.path.exists(hours_added_path):
    os.makedirs(hours_added_path)  
hours_added_path_boot = '/content/drive/MyDrive/cpbandit/CP_Bandit/data/hours2sepsis_boots/'
if not os.path.exists(hours_added_path_boot):
    os.makedirs(hours_added_path_boot)  
data_path = '/content/drive/MyDrive/cpbandit/CP_Bandit/data/all_dataset'
```

```python
src_allfiles = os.listdir(data_path)
all_dfs = []
```

```python
import math
bootstep = 500
num_boot = int(math.floor(len(src_allfiles)/500))
print(bootstep,num_boot)
```

```python
len(src_allfiles)
```

```python
len(src_allfiles)%bootstep
```

```python
len(src_allfiles)%num_boot
```

```python
import time
start_time = time.time()
for i in range(num_boot+1):
  start_time_boot = time.time()
  # the index of list 'src_allfiles' must be integer
  istart = int(i*bootstep)
  if i == num_boot:
    istop = int(istart)+int(len(src_allfiles)%bootstep)-1
  else:
    istop = int((i+1)*bootstep-1)
  print('Processing Boot {}-th ... '.format(i))
  print('Patients Index from {} to {}'.format(istart,istop))

  temp_dfs = []
  
  for psv in src_allfiles[istart:istop]:
      curr_path = os.path.join(data_path,psv)
      # header=0, after reading the CSV file's first row,
      # it can be assigned as the column names.
      curr_psv = pd.read_csv(curr_path, index_col = None, header = 0, sep='|')
      curr_psv['pat_id'] = psv.split(".")[0]

      if 'SepsisLabel' not in curr_psv.columns:
          pass

      else:
          if 1 in np.array(curr_psv.SepsisLabel) and curr_psv.SepsisLabel.ne(0).idxmax()==0:
            curr_psv['hours2sepsis'] =0 

          elif 1 in np.array(curr_psv.SepsisLabel) and curr_psv.SepsisLabel.ne(0).idxmax()!=0:
              # ne(0): not equal to zero
              # idxmax() gives you the first position of the maximum value
              # ignore the six hours issue.
              t_sepsis= curr_psv.SepsisLabel.ne(0).idxmax()+6 
              hours2sepsis = list(reversed(range(t_sepsis+1))) + [0] * (len(curr_psv)-t_sepsis-1)

              if len(hours2sepsis) != len(curr_psv):
                hours2sepsis = list(reversed(range(t_sepsis-len(curr_psv)+1, t_sepsis+1))) 
                print(psv, ':different length')
              curr_psv['hours2sepsis'] = hours2sepsis
                # break              
          else: # assign a very big number for the nonseptic patients       
              curr_psv['hours2sepsis'] = 500 
          curr_psv.to_csv(os.path.join(hours_added_path,psv))  
          all_dfs.append(curr_psv)
          temp_dfs.append(curr_psv)
  print('Boot:', i,'done!')
  print("--- %s seconds used for relabeling boot {}--- ".format(i) % (time.time() - start_time_boot))
  tempdata_csv= pd.concat(temp_dfs, ignore_index=True)
  tempdata_csv.to_csv('/content/drive/MyDrive/cpbandit/CP_Bandit/data/hours2sepsis_boots/boot_{}'.format(i))
allinone_csv = pd.concat(all_dfs, ignore_index=True)
allinone_csv.to_csv('/content/drive/MyDrive/cpbandit/CP_Bandit/data/allin1.csv') 
print("--- %s seconds used for relabeling hours to sepsis--- " % (time.time() - start_time))
```
